## Outlier Analysis

Outliers ile mucadele de genel olarak 3 yaklasim vardir.
* Aykırı değerleri veri kümemizden çıkarabiliriz.
* Aykırı değerlere yeni değerler atayarak sınırlayabiliriz.
* Onları analizimiz için zararsız bir değere dönüştürürüz.

### Functions

### Tukey’s Method
* Bu yöntem IQR yöntemi olarak da bilinir.
* Bu yöntemde IQR in 1.5 katı dışında olan tüm değerleri outlier değerler olarak değerlendiririz. 
* Ancak, aralığı değiştirebiliriz(Q1 icin 0.25 degilde 0.1 alma gibi) ve 1.5 yerine başka bir eşik değeri ayarlayabiliriz.

image.png

### Z-Score

* Bir değişkende aykırı değerleri saptama yöntemlerinden birisi “z-score” dur. * Z-score ilk önce verileri ölçeklendirir, böylece ortalaması 0 ve standart sapması 1 olur ve daha sonra bir değerin ortalamanın ne kadar uzağında olduğunu ölçer. 
* Bununla birlikte, bir değerin aykırı olduğunu söylemek için, z-score için bir eşik belirlememiz gerekir, bu sayede bu eşiğin üzerindeki skorların aykırı olduğu söylenir. 
* Çoğu durumda 3 veya -3 sigma eşiği kullanılır. Z-score değeri sırasıyla 3 veya -3'ten büyük veya küçükse, bu veri noktası aykırı değer olarak belirlenecektir.

## REGEX

https://regex101.com/

\d Any numeric digit from 0 to 9.

\D Matches any character which is not a decimal digit. This is the opposite of \d.

\w Any letter, numeric digit, or the underscore character. (Think of this as matching "word" characters.)

\W Any character that is not a letter, numeric digit, or the underscore character.

\s Any space, tab, or newline character. ( Think of this as matching white-space characters.)

\S Any character that is not a space, tab, or newline.

### Special Character

" [ ] " A set of characters "[a-z]"

" \ " Signals a special sequence (can also be used to escape special characters)

" . " Any character (except newline character)

" ^ " Starts with "^hello"

" \$ " Ends with "world\$"

" * " Zero or more occurrences

" + " One or more occurrences

" { } " Exactly the specified number of occurrences

" | " Either or "falls|stays"

" ( ) " Capture and group
